```python
# Step 2: 데이터 불러오기
import pandas as pd
file_path = "/content/drive/MyDrive/dataset/미세먼지_전국지점_일별_202503.csv"
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
print(df.shape)
```

```python
# Step 3: 필요한 컬럼만 추출
# 컬럼명은 실제 데이터에 맞게 조정 필요
df = df[['지점명', '일 미세먼지 농도(㎍/㎥)']]  # 예: '서울', '부산' 등 지역, PM10 농도 수치
df = df.dropna()
```

```python
# Step 4: 한글 시각화를 위한 라이브러리
!pip install koreanize-matplotlib
import matplotlib.pyplot as plt
import koreanize_matplotlib

# Step 5: 박스플롯 시각화
plt.figure(figsize=(20, 6))
region_list = df['지점명'].unique()
data = [df[df['지점명'] == region]['일 미세먼지 농도(㎍/㎥)'] for region in region_list]

plt.boxplot(data, labels=region_list)
plt.title("지역별 미세먼지(PM10) 농도 분포 (박스플롯)")
plt.ylabel("일 미세먼지 농도 (㎍/㎥)")
plt.xlabel("지역")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
